# Filling Missing Values

```python
#import Library
import pandas as pd
```

```python
# Get cleaned data
dataframe = pd.read_csv('Cleaned_covid_data.csv')
```

```python
# Fill all columns given below with the mode value of the same column grouping them by age 
columns_to_fill = [
    'PNEUMONIA', 'PREGNANT', 'DIABETES', 'ASTHMA', 'INMSUPR', 'HIPERTENSION',
    'CARDIOVASCULAR', 'COPD', 'RENAL_CHRONIC', 'OBESITY', 'TOBACCO', 'INTUBED','ICU']

# Fill missing values in columns with mode grouped by 'AGE' or default value (0)
for col in columns_to_fill:
    mode_values = dataframe.groupby('AGE')[col].apply(lambda x: x.mode())
    for age, mode_val in mode_values.iteritems():
        if pd.notnull(mode_val) and dataframe['AGE'].isin([age]).any():
            dataframe.loc[dataframe['AGE'] == age, col] = dataframe.loc[dataframe['AGE'] == age, col].fillna(mode_val.iloc[0])
        else:
            dataframe[col] = dataframe[col].fillna(0)
```

```python
# make a copy of all rows where the CLASIFFICATION_FINAL is null which will be used to fill after training the ML model
df_null_classification = dataframe[dataframe['CLASIFFICATION_FINAL'].isnull()].copy()
dataframe.dropna(subset=['CLASIFFICATION_FINAL'], inplace=True)
```

```python
dataframe.to_csv('Covid_Severity_train_data.csv', index = False)
df_null_classification.to_csv('Covid_Severity_test_data.csv', index = False)
```
